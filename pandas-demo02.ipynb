## How to handle time series data with ease

```python
air_quality = pd.read_csv("~/pandas-tutorials/air_quality_no2_long.csv")
air_quality.head()
```

```python
air_quality.city.unique()
```

# Columns Rename

```python
air_quality = air_quality.rename(columns={"date.utc": "datetime"})
air_quality.head()
```

# Using pandas datetime properties

```python
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
air_quality["datetime"]
```

```python
air_quality["datetime"].min(), air_quality["datetime"].max()
```

```python
air_quality["datetime"].max() - air_quality["datetime"].min()
```

# Add a new column to the DataFrame containing only the month of the measurement

```python
air_quality["month"] = air_quality["datetime"].dt.month
air_quality.head(10)
```

```python
air_quality["weekday"] = air_quality["datetime"].dt.weekday
air_quality.head()
```

```python
air_quality.groupby([air_quality["datetime"].dt.month, "location"])["value"].mean()
```

```python
air_quality.groupby([air_quality["datetime"].dt.weekday, "location"])["value"].mean()
```

# Plot the typical pattern during the day of our time series of all stations together

```python
fig, axs = plt.subplots(figsize=(12,4))
air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(kind='bar', rot=0, ax=axs)
plt.xlabel("Hour of the day");
plt.ylabel("$NO_2 (Âµg/m^3)$");  # custom x label using Matplotlib
```

# Datetime as index

```python
no_2 = air_quality.pivot(index="datetime", columns="location", values="value")
no_2.head()
```

```python
no_2.index.year, no_2.index.weekday
```

```python
no_2["2019-05-20":"2019-05-21"].plot();
```

# Resample a time series to another frequency

```python
monthly_max = no_2.resample("5H").max()
monthly_max
```

```python
monthly_max.index.freq
```

```python
no_2.resample("D").mean().plot(style="-o", figsize=(10, 5));
```
